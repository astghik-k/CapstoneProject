```python
import os
import cv2
import numpy as np

def rgb_to_hsv(rgb_color):
    """Convert an RGB color to HSV using OpenCV."""
    color_bgr = np.uint8([[list(reversed(rgb_color))]])  # Convert RGB to BGR
    color_hsv = cv2.cvtColor(color_bgr, cv2.COLOR_BGR2HSV)
    return color_hsv[0][0]

def mask_image(img_src, lower, upper):
    """Convert image from BGR to HSV and create a mask for given lower and upper boundaries."""
    img_hsv = cv2.cvtColor(img_src, cv2.COLOR_BGR2HSV)
    hsv_lower = np.array(lower, np.uint8)
    hsv_upper = np.array(upper, np.uint8)
    img_mask = cv2.inRange(img_hsv, hsv_lower, hsv_upper)
    return img_mask, img_hsv

def apply_removal_operations(image_path, output_path):
    img_src = cv2.imread(image_path)
    # Removing red marks
    hsv = cv2.cvtColor(img_src, cv2.COLOR_BGR2HSV)
    lower_red1 = np.array([0, 30, 30])
    upper_red1 = np.array([10, 255, 255])
    lower_red2 = np.array([170, 50, 50])
    upper_red2 = np.array([180, 255, 255])
    mask1 = cv2.inRange(hsv, lower_red1, upper_red1)
    mask2 = cv2.inRange(hsv, lower_red2, upper_red2)
    red_mask = cv2.bitwise_or(mask1, mask2)
    red_coords = np.column_stack(np.where(red_mask > 0))
    for coord in red_coords:
        cv2.circle(img_src, (coord[1], coord[0]), 10, (255, 255, 255), -1)

    # Removing yellow highlights
    highlight_rgb_yellow = [254, 254, 164]  # Yellow highlighter color
    highlight_hsv_yellow = rgb_to_hsv(highlight_rgb_yellow)
    hsv_range = 10
    lower_hsv_yellow = [highlight_hsv_yellow[0] - hsv_range, 100, 100]
    upper_hsv_yellow = [highlight_hsv_yellow[0] + hsv_range, 255, 255]
    yellow_mask, _ = mask_image(img_src, lower_hsv_yellow, upper_hsv_yellow)
    kernel = np.ones((40, 40), np.uint8)  # Adjust the dilation radius as needed
    yellow_mask_dilated = cv2.dilate(yellow_mask, kernel, iterations=1)
    img_src[yellow_mask_dilated == 255] = (255, 255, 255)

    # Removing blue highlights
    highlight_rgb_blue = [0, 0, 255]  # Blue highlighter color (in BGR format)
    highlight_hsv_blue = rgb_to_hsv(highlight_rgb_blue)
    lower_hsv_blue = [highlight_hsv_blue[0] - hsv_range, 100, 100]
    upper_hsv_blue = [highlight_hsv_blue[0] + hsv_range, 255, 255]
    blue_mask, _ = mask_image(img_src, lower_hsv_blue, upper_hsv_blue)
    blue_mask_dilated = cv2.dilate(blue_mask, kernel, iterations=1)
    img_src[blue_mask_dilated == 255] = (255, 255, 255)

    # Removing orange highlights
    highlight_rgb_orange = [0, 165, 255]  # Orange highlighter color (in BGR format)
    highlight_hsv_orange = rgb_to_hsv(highlight_rgb_orange)
    lower_hsv_orange = [highlight_hsv_orange[0] - hsv_range, 100, 100]
    upper_hsv_orange = [highlight_hsv_orange[0] + hsv_range, 255, 255]
    orange_mask, _ = mask_image(img_src, lower_hsv_orange, upper_hsv_orange)
    orange_mask_dilated = cv2.dilate(orange_mask, kernel, iterations=1)
    img_src[orange_mask_dilated == 255] = (255, 255, 255)
    cv2.imwrite(output_path, img_src)

def process_images_in_folder(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(input_folder):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, filename)
            apply_removal_operations(image_path, output_path)
            print("Processed:", filename)

# Example usage
input_folder = 'handwriting'
output_folder = 'processed'

process_images_in_folder(input_folder, output_folder)
```
