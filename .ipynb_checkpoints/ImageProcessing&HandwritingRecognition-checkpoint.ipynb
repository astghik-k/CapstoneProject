## Import necessary libraries

```python
import os
import io
import cv2
import openpyxl
import numpy as np
import pandas as pd
from PIL import Image
import seaborn as sns
from google.cloud import vision
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from skimage import io, img_as_ubyte
from pdf2image import convert_from_path
from scipy.ndimage import gaussian_filter
from google.oauth2 import service_account
from skimage.filters import threshold_otsu
from ast import literal_eval as safe_literal_eval
```

## Image processing functions

```python
def process_image(image_path):
    """
    Process the image to find contours and the largest ellipse. It returns the processed image path
    and the x-coordinates of the vertical lines that would be drawn at the extremities of the largest ellipse.
    """
    
    image = io.imread(image_path)
    gray_image = rgb2gray(image)
    thresh = threshold_otsu(gray_image)
    binary = gray_image > thresh
    binary_uint8 = img_as_ubyte(binary)
    blurred = gaussian_filter(binary_uint8, sigma=10)
    thresh_blurred = threshold_otsu(blurred)
    binary_blurred = blurred > thresh_blurred
    binary_blurred_uint8 = img_as_ubyte(binary_blurred)
    kernel = np.ones((3, 3), np.uint8)
    filled = cv2.morphologyEx(binary_blurred_uint8, cv2.MORPH_CLOSE, kernel)
    eroded = cv2.erode(filled, kernel, iterations=1)
    contours_eroded, _ = cv2.findContours(eroded, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    area_threshold = (eroded.shape[0] * eroded.shape[1]) / 2

    image_with_vertical_ellipses = cv2.cvtColor(eroded, cv2.COLOR_GRAY2BGR)
    max_area = 0
    max_ellipse_bbox = None
    for cnt in contours_eroded:
        if cv2.contourArea(cnt) < area_threshold and len(cnt) >= 5:
            ellipse = cv2.fitEllipse(cnt)
            area = cv2.contourArea(cnt)
            if area > max_area:
                max_area = area
                max_ellipse_bbox = cv2.boundingRect(cnt)

    # Initializing borders to the image width in case no contour is large enough
    x1, x2 = 0, image_with_vertical_ellipses.shape[1]

    if max_ellipse_bbox:
        x, y, w, h = max_ellipse_bbox
        x1 = x
        x2 = x + w

    output_path = 'image_with_largest_ellipse_and_lines.jpg'
    cv2.imwrite(output_path, image_with_vertical_ellipses)

    return output_path, x1, x2


def merge_large_ellipses(image_path):
    """
    Merge contours that represent large ellipses in the image based on a dynamic area threshold.
    This function is used to identify the bottom-most y-coordinate of these ellipses for further processing.
    Returns the bottom-most y-coordinate of the merged large ellipses or zero if no large ellipses are found.
    """
    
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        print("Error: Image cannot be loaded. Check the path or file.")
        return None

    # Adaptive threshold might be more suitable depending on image variety
    _, binary = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Dynamic threshold based on image size
    area_threshold = img.size * 0.0107  # 1% of image area
    large_contours = [cnt for cnt in contours if cv2.contourArea(cnt) > area_threshold]

    text_mask = np.zeros_like(binary)
    cv2.drawContours(text_mask, large_contours, -1, color=255, thickness=-1)
    kernel = np.ones((50, 50), np.uint8)  # Consider adjusting the kernel size
    dilated_mask = cv2.dilate(text_mask, kernel, iterations=1)
    bottom_most_y = max([cv2.boundingRect(c)[1] + cv2.boundingRect(c)[3] for c in large_contours], default=0)

    return bottom_most_y


def crop_and_save_other_half(input_path, bottom):
    """
    Crop the image below the identified bottom-most point and save the cropped image
    in a specific folder.
    """
    image = cv2.imread(input_path)

    if image is None:
        print("Error: Image cannot be loaded. Check the path or file.")
        return

    cropped_img = image[bottom:, :]
    pics_dir = 'handwriting'
    os.makedirs(pics_dir, exist_ok=True)
    cropped_img_name = os.path.splitext(os.path.basename(input_path))[0] + '_cropped.jpg'
    cropped_img_path = os.path.join(pics_dir, cropped_img_name)
    cv2.imwrite(cropped_img_path, cropped_img)

    return cropped_img_path


def process_all_images(images_dir):
    """
    Process all images in the specified directory.
    Each image will be processed to detect and isolate text blobs with large ellipses,
    then crop the image to focus on the handwriting part, and save the cropped images
    in a 'handwriting' folder.
    Args:
    - images_dir (str): The directory containing the images to process.
    """
   
    handwriting_dir = 'handwriting'
    os.makedirs(handwriting_dir, exist_ok=True)


    image_files = [f for f in os.listdir(images_dir) if os.path.isfile(os.path.join(images_dir, f))]

    for image_file in image_files:
        # Construct the full path to the image file
        image_path = os.path.join(images_dir, image_file)
        print(f"Processing {image_path}...")

        # Process the image
        try:
            processed_image_path, x1, x2 = process_image(image_path)
            bottom = merge_large_ellipses(processed_image_path)
            cropped_image_path = crop_and_save_other_half(image_path, bottom)
            print(f"Cropped image saved at: {cropped_image_path}")
        except Exception as e:
            print(f"Error processing {image_path}: {e}")


def convert_pdfs_to_jpgs(pdf_folder, output_folder):
    """
    Convert all PDF files found in a specified directory to JPG images.
    """
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    pdf_files = [f for f in os.listdir(pdf_folder) if f.endswith('.pdf')]
    for pdf_file in pdf_files:
        pdf_path = os.path.join(pdf_folder, pdf_file)
        images = convert_from_path(pdf_path)
        for page_number, image in enumerate(images, start=1):
            image_filename = f"{os.path.splitext(pdf_file)[0]}_page_{page_number}.jpg"
            image_path = os.path.join(output_folder, image_filename)
            image.save(image_path, 'JPEG')


def rgb_to_hsv(rgb_color):
    """
    Convert an RGB color to HSV using OpenCV.
    """
    color_bgr = np.uint8([[list(reversed(rgb_color))]]) 
    color_hsv = cv2.cvtColor(color_bgr, cv2.COLOR_BGR2HSV)
    return color_hsv[0][0]


def process_color_removal(img_src, lower, upper):
    """
    Apply color removal using morphological operations on detected areas.
    """
    hsv = cv2.cvtColor(img_src, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv, np.array(lower, np.uint8), np.array(upper, np.uint8))
    kernel = np.ones((50, 50), np.uint8)  
    closed_mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
    contours, _ = cv2.findContours(closed_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    for contour in contours:
        if cv2.contourArea(contour) > 50:  
            M = cv2.moments(contour)
            if M["m00"] != 0:
                cx = int(M["m10"] / M["m00"])
                cy = int(M["m01"] / M["m00"])
                top_left_x = max(cx - 290, 0)
                top_left_y = max(cy - 30, 0)
                bottom_right_x = min(cx + 290, img_src.shape[1])
                bottom_right_y = min(cy + 30, img_src.shape[0])
                cv2.rectangle(img_src, (top_left_x, top_left_y), (bottom_right_x, bottom_right_y), (255, 255, 255), -1)

def apply_removal_operations(image_path, output_path):
    """
    Read an image from a file path, apply color removal operations, and save the processed image to another path.
    """
    img_src = cv2.imread(image_path)
    if img_src is None:
        print("Error: Image cannot be loaded. Check the path or file.")
        return
    
    if img_src.shape[1] < 200 or img_src.shape[0] < 200:
        print("Skipping processing due to image size:", image_path)
        return

    # Process highlighter colors
    highlight_hsv_yellow = rgb_to_hsv([255, 255, 130])
    highlight_hsv_blue = rgb_to_hsv([135, 206, 250])
    highlight_hsv_orange = rgb_to_hsv([255, 165, 0])
    process_color_removal(img_src, [highlight_hsv_yellow[0] - 10, 100, 100], [highlight_hsv_yellow[0] + 10, 255, 255])
    process_color_removal(img_src, [highlight_hsv_blue[0] - 10, 100, 100], [highlight_hsv_blue[0] + 10, 255, 255])
    process_color_removal(img_src, [highlight_hsv_orange[0] - 10, 100, 100], [highlight_hsv_orange[0] + 10, 255, 255])

    cv2.imwrite(output_path, img_src)

def process_images_in_folder(input_folder, output_folder):
    """
    Process all image files in the specified input folder, apply removal operations, and save them in the specified output folder.
    """
    
    for filename in os.listdir(input_folder):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, filename)
            apply_removal_operations(image_path, output_path)
            print("Processed:", filename)

```

## Handwriting recognition functions

```python
def authenticate_with_service_account(key_path):
    """
    Authenticate to a Google Cloud service account using a provided key file path.
    """
    credentials = service_account.Credentials.from_service_account_file(key_path)
    client = vision.ImageAnnotatorClient(credentials=credentials)
    return client


def detect_handwritten_text(image_path, client, visually_similar_groups):
    """
    Detects handwritten text within an image and evaluates the confidence of character recognition,
    including an evaluation of visually similar characters.
    """

    with open(image_path, 'rb') as image_file:
        content = image_file.read()
    image = vision.Image(content=content)
    image_context = vision.ImageContext(language_hints=['en', 'el'])
    response = client.document_text_detection(image=image, image_context=image_context)
    document = response.full_text_annotation
    full_text = document.text
    detailed_confidences = []
    character_confidences = {}
    for page in document.pages:
        for block in page.blocks:
            for paragraph in block.paragraphs:
                for word in paragraph.words:
                    for symbol in word.symbols:
                        char = symbol.text
                        confidence = symbol.confidence
                        character_confidences[char] = confidence
                        similar_chars_confidences = {}
                        if confidence < 1:
                            for group, similar_chars in visually_similar_groups.items():
                                if char in similar_chars:
                                    for similar_char in similar_chars:
                                        if similar_char != char and similar_char in character_confidences:
                                            similar_chars_confidences[similar_char] = character_confidences[similar_char]
                        detailed_confidences.append({
                            'image_name': os.path.basename(image_path),
                            'character': char,
                            'confidence': confidence,
                            'similar_chars': similar_chars_confidences if similar_chars_confidences else None
                        })
    return full_text, detailed_confidences


def process_folder(folder_path, key_path, output_text_file, output_excel_file):
    """
    Processes all images in a folder to detect handwritten text, saves the full text to a text file,
    and exports detailed confidence data for characters to an Excel file.
    """
    
    client = authenticate_with_service_account(key_path)
    visually_similar_groups = {
        'Group 1': {'(', 'c'},
        'Group 2': {'c', '<'},
        'Group 3': {'c', 'e'},
        'Group 4': {'e', 'l'},
        'Group 5': {'e', 'o'},
        'Group 6': {'o', '0'},
        'Group 7': {'=', 'z'},
        'Group 8': {'z', 'r'},
        'Group 9': {'r', '2'},
        'Group 10': {'z', '2'},
        'Group 11': {'v', 'r'},
        'Group 12': {'r', 'n'},
        'Group 13': {'t', '+'},
        'Group 14': {'y', 'g'},
        'Group 15': {'j', 'i'},
        'Group 16': {'i', ';'},
        'Group 17': {'j', ';'},
        'Group 18': {'{', '('},
        'Group 19': {'O', 'D'}}
    all_confidences = []
    with open(output_text_file, 'w') as text_file:
        for filename in os.listdir(folder_path):
            if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
                image_path = os.path.join(folder_path, filename)
                full_text, image_confidences = detect_handwritten_text(image_path, client, visually_similar_groups)
                text_file.write(f"{filename}: {full_text}\n")
                all_confidences.extend(image_confidences)
    df = pd.DataFrame(all_confidences)
    df.to_excel(output_excel_file, index=False)
    
def extract_letter(image_name):
    """
    Extracts a letter code from an image filename, which indicates a particular label or category.
    """
    parts = image_name.split('_')
    for i, part in enumerate(parts):
        if 'page' in part:
            code_part = parts[i-1]
            return code_part[-4] if code_part[-4] in ['H', 'M', 'L'] else ''
    return ''

def create_excel_from_images(directory):
    """
    Creates an Excel file from image filenames in a directory, extracting labels from filenames and storing them.
    """
data = []
    
    for filename in os.listdir(directory):
        if filename.endswith(".jpg") or filename.endswith(".png"):  # check for image files
            label = extract_letter(filename)
            data.append({'image_name': filename, 'label': label})

    df = pd.DataFrame(data)
    df.to_excel('image_labels.xlsx', index=False)
```
